```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import K_clust
def normalise(Xin):
        return (Xin-np.mean(Xin,axis=0))/np.std(Xin,axis=0)    # NORMALISING FEATURES

data=np.genfromtxt('Image1.txt',delimiter=',')
X=data[:,:-1]
X=normalise(X)   #normalising X
plt.scatter(X[:,0],X[:,1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Training Data")
plt.show()
```

```python
train=K_clust.K_clust()
train.set(X,noc=2)  # For 2 Clusters
prediction=train.fit(noi=50)
```

```python
plt.scatter(X[:,0],X[:,1],c=prediction)
plt.title('My Predicted Clusters')
plt.show()
```

<<=====Comparing my predictions with that of SKlearn=====>>

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2, random_state=0)
model.fit(X)
predicted=model.predict(X)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=predicted)
plt.title('Test Set (as per SK Learn)')
plt.show()
```
